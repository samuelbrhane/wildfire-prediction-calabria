```python
# Imports
import os
import pandas as pd
```

```python
# Load and merge all yearly CSVs into one DataFrame
data_folder = "../1_data/processed/yearly"
csv_files = [f for f in os.listdir(data_folder) if f.endswith(".csv")]

all_data = []
for file in csv_files:
    df = pd.read_csv(os.path.join(data_folder, file))
    df["Date"] = pd.to_datetime(df["Date"], format="%Y%m%d")
    all_data.append(df)

climate_fire_df = pd.concat(all_data, ignore_index=True)
climate_fire_df.sort_values(["Zone_ID", "Date"], inplace=True)
climate_fire_df
```

```python
# Clean and save merged dataset
climate_fire_df["Zone_ID"] = climate_fire_df["Zone_ID"].astype(int)
climate_fire_df.reset_index(drop=True, inplace=True)

columns = ["Date"] + [col for col in climate_fire_df.columns if col != "Date"]
climate_fire_df = climate_fire_df[columns]

climate_fire_df.to_csv("../1_data/processed/zone_sequence_merged.csv", index=False)
print(f"Merged dataset saved: {climate_fire_df.shape[0]} rows, {climate_fire_df.shape[1]} columns")
climate_fire_df
```
